# Evaluation of the Summary Documents

This notebook generates the Rouge matrix for the summary documents. We are using sacrerouge module to generate this  matrix.

```python
import os
import json
from sacrerouge.metrics import Rouge
metric = Rouge()


goldlist=[]

#BASELINE
#basepath="/Users/dhavaldangaria/Documents/Dhaval/stanford/cs224n/Project/sumo/results/output_eval_full"

#BASEBERT
#basepath="/Users/dhavaldangaria/Documents/Dhaval/stanford/cs224n/Project/bert_executivesummary/bert-extractive-summarization/results_full"

#BASEBIGBIRD
basepath="/Users/dhavaldangaria/Documents/Dhaval/stanford/cs224n/Project/sumo/results/eval_doc_summary_BigBirdv2/eval_sect_summary_ckpt1296_512_v6"
basepathgold="/Users/dhavaldangaria/Documents/Dhaval/stanford/cs224n/Project/sumo/eval_articles/validation_gold_standards/"
 

for filename in os.listdir(basepathgold):
    if filename.endswith(".txt"): 
        goldlist.append(filename)
#print(goldlist)


basepath=basepath+'/'

count=1
summary=[]
ref=[]
for filename in os.listdir(basepath):
    if filename.endswith(".txt"): 
        d=[]
        
        if "_"in filename:
            filestr=filename.split("_")
            for fn in goldlist:
                if fn.startswith(filestr[0]):
                    with open(basepathgold+fn, 'r') as file:
                        data = file.read()
                        d.append(data)
            with open(basepath+filename, 'r') as file:
                        data = file.read()
            summary.append(data) 
            ref.append(d)
            #break
            
print(len(summary))
print(len(ref))                   
scores = metric.evaluate(summary, ref)
print(scores)
            
            
          

        
```
